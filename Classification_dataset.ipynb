In this file,we have worked on getting the final attributes required for training the model

```python
import pandas as pd
```

```python
df = pd.read_parquet('preprocessed_parquet',engine='fastparquet')
df.head()
```

```python
req_cols = ['num_of_prev_attempts','studied_credits','date_registration','sum_click','date_assessment','assessment_type','weight','score']
```

```python
df1 = pd.DataFrame()
for i in req_cols:
    df1[i] = df[i]
```

```python
df1.head(10)
```

```python
df1.shape
```

Dropping hte duplicates 

```python
df1 = df1.drop_duplicates()
```

```python
df1.shape
```

The score is encoded to the required classes

```python
def score_encoding(i):
    if i>=0 and i<=10:
        return 0
    if i>=11 and i<=20:
        return 1
    if i>=21 and i<=30:
        return 2
    if i>=31 and i<=40:
        return 3
    if i>=41 and i<=50:
        return 4
    if i>=51 and i<=60:
        return 5
    if i>=61 and i<=70:
        return 6
    if i>=71 and i<=80:
        return 7
    if i>=81 and i<=90:
        return 8
    if i>=91 and i<=100:
        return 9
```

```python
new_score = []
for i in df1['score']:
    new_score.append(score_encoding(i))
```

```python
df1['score'] = new_score
```

```python
df1['score'].unique()
```

One-hot encoding the columns

```python
df1  = pd.get_dummies(df1, columns = ['assessment_type'])
```

```python
df1.sample(10)
```

Gettting the final file for trraining the model

```python
df1.to_parquet('classification_parquet',index=False)
```
